```python
# Calculate Riemann sum of a function
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**2

# Define the interval
a = 0
b = 1

# Define the number of points
n = 100

# Define the step size
h = (b - a) / n

# Define the sum
sum = 0

# Loop through the interval
for i in range(n):
    # Calculate the x value
    x = a + i * h
    # Calculate the area
    area = h * (f(x) + f(x + h)) / 2
    # Add the area to the sum
    sum += area

    # Print the area
    print("Area:", area)
    print("Sum:", sum)


# Plot the function 
x = np.linspace(a, b, n)
y = f(x)
plt.plot(x, y)
plt.show()
```

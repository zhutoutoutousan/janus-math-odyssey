```python
# Calculate multi-variable integral of the function f(x, y) = x^2 + y^2
from scipy.integrate import dblquad
definite_integral = dblquad(lambda x, y: x**2 + y**2, 0, 1, lambda x: 0, lambda x: 1)
# plot from -100 to 100
x = np.linspace(-100, 100, 1000)
y = np.linspace(-100, 100, 1000)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
plt.contour(X, Y, Z)
# legend
plt.legend(['f(x, y) = x^2 + y^2'])
# title
plt.title('Definite integral of f(x, y) = x^2 + y^2')
# Color the regions of the contour map that correspond to the integral of the function
plt.contourf(X, Y, Z, cmap='viridis')
# Show colorbar for the contour map
plt.colorbar()
# x-axis label
plt.xlabel('x')
# y-axis label
plt.ylabel('y')
# show plot
plt.show()
```

```python
# Calculate multi-variable integral of the function f(x, y, z) = x^2 + y^2 + z^2
from scipy.integrate import tplquad
definite_integral = tplquad(lambda x, y, z: x**2 + y**2 + z**2, 0, 1, lambda x: 0, lambda x: 1, lambda y: 0, lambda y: 1)
# plot from -100 to 100
x = np.linspace(-100, 100, 1000)
y = np.linspace(-100, 100, 1000)
z = np.linspace(-100, 100, 1000)
X, Y, Z = np.meshgrid(x, y, z)
W = X**2 + Y**2 + Z**2
plt.contour(X, Y, W)
# legend
plt.legend(['f(x, y, z) = x^2 + y^2 + z^2'])
# title
plt.title('Definite integral of f(x, y, z) = x^2 + y^2 + z^2')
# Color the regions of the contour map that correspond to the integral of the function
plt.contourf(X, Y, W, cmap='viridis')
# Show colorbar for the contour map
plt.colorbar()
# x-axis label
plt.xlabel('x')
# y-axis label
plt.ylabel('y')
# show plot
plt.show()
```

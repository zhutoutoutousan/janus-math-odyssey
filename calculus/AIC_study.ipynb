```python
import numpy as np
import matplotlib.pyplot as plt

# Statistics 101: Multiple Regression, AIC, AICc, and BIC Basics: https://www.youtube.com/watch?v=-BR4WElPIXg&t=1s
# https://www.tutorialspoint.com/how-to-plot-a-multivariate-function-in-python-matplotlib
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

def func(x, y):
   return 3 * x + 4 * y - 2 + np.random.randn(30)

x, y = np.random.randn(2, 30)
y *= 100
z = func(x, y)

fig, ax = plt.subplots()

s = ax.scatter(x, y, c=z, s=100, marker='*', cmap='plasma')

fig.colorbar(s)

plt.show()
```
